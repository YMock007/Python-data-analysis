```python
import numpy as np

# Importing Data from intake and Enrolment
data_intake = np.genfromtxt('IntakebyInstitutions.csv', delimiter=",", names=True,
                            dtype=['int', 'U50', 'int','int','int','int','int','int','int','int','int','int','int','int','int','int','int','int','int'])

data_enrolment = np.genfromtxt('EnrolmentbyInstitutions.csv', delimiter=",", names=True,
                               dtype=['int', 'U50', 'int','int','int','int','int','int','int','int','int','int','int','int','int','int','int','int','int'])

data_graduate = np.genfromtxt('Graduatesbyinstitutions.csv', delimiter=",", names=True,
                               dtype=['int', 'U50', 'int','int','int','int','int','int','int','int','int','int','int','int','int','int','int','int','int'])

print("Total number of rows in EnrolmentbyInstitutions.csv :", data_intake.shape[0])
print("Total number of rows in IntakebyInstitutions.csv :", data_enrolment.shape[0])
print("Total number of rows in GraduatesbyInstitutions.csv :", data_graduate.shape[0])
period = np.unique(data_intake['year'])
print(f"Every CSV files have the time period from {period[0]} to {period[-1]}")
print('There are 16 different institutions.')

mask_data_intake = (data_intake['year'] >= 2014) & (data_intake['year'] <= 2021) & (data_intake['sex'] == 'MF')
required_data_intake = np.array([data_intake[mask_data_intake]["nus"],
                                data_intake[mask_data_intake]["ntu"],
                                data_intake[mask_data_intake]["smu"],
                                data_intake[mask_data_intake]["sit"],
                                data_intake[mask_data_intake]["sutd"],
                                data_intake[mask_data_intake]["suss"]])

mask_data_enrolment = (data_enrolment['year'] >= 2014) & (data_enrolment['year'] <= 2021) & (data_enrolment['sex'] == 'MF')
required_data_enrolment = np.array([data_enrolment[mask_data_enrolment]["nus"],
                                    data_enrolment[mask_data_enrolment]["ntu"],
                                    data_enrolment[mask_data_enrolment]["smu"],
                                    data_enrolment[mask_data_enrolment]["sit"],
                                    data_enrolment[mask_data_enrolment]["sutd"],
                                    data_enrolment[mask_data_enrolment]["suss"]])

mask_data_graduate = (data_graduate['year'] >= 2014) & (data_graduate['year'] <= 2021) & (data_graduate['sex'] == 'MF')
required_data_graduate = np.array([data_graduate[mask_data_graduate]["nus"],
                                    data_graduate[mask_data_graduate]["ntu"],
                                    data_graduate[mask_data_graduate]["smu"],
                                    data_graduate[mask_data_graduate]["sit"],
                                    data_graduate[mask_data_graduate]["sutd"],
                                    data_graduate[mask_data_graduate]["suss"]])

total_intake = required_data_intake.sum(axis=0)
total_enrollment = required_data_enrolment.sum(axis=0)
total_graduates = required_data_graduate.sum(axis=0)
print(f"{'='*40}")
print(f"{'Total Students in Singapore''s Universities':^40}")
print(f"{'='*40}")
print(f"The total number of students enrolled: {total_enrollment.sum():,}")
print(f"The total number of students admitted: {total_intake.sum():,}")
print(f"The total number of students graduated: {total_graduates.sum():,}")
print(f"{'='*40}")

universities = ["NUS", "NTU", "SMU", "SIT", "SUTD", "SUSS"]
# Calculate totals for each university
total_intake_per_university = required_data_intake.sum(axis=1)
total_enrollment_per_university = required_data_enrolment.sum(axis=1)
total_graduates_per_university = required_data_graduate.sum(axis=1)

print(total_enrollment )
for i, university in enumerate(universities):
    print(university)
    print(total_intake[i])
    print(total_enrollment[i])
    print(total_graduates[i])
    print('')
```
